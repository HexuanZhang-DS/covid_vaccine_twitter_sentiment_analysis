# Query by Text Search

# Function to get tweets

```python
def get_tweets(tweet_count, text_query, since_date, until_date):
    """
    This is a function to use snsscrape to get tweets based on four parameters:
    tweet_count: number of tweets scraped per day
    text_query: a string to search targetted tweets
    since_date: starting date for the scraping
    until_date: ending date for the scraping
    
    """
    since_date=pd.to_datetime(since_date)
    until_date=pd.to_datetime(until_date)
    dates= pd.date_range(start=since_date, end=until_date).strftime("%Y-%m-%d")
    i=0
    while i <len(dates)-1:
        start = dates[i]
        end = dates[i+1]
        os.system('snscrape --jsonl --max-results {} --since {} twitter-search "{} until:{}"> text-query-tweets_{}.json'.format(tweet_count, start, text_query, end, end))
        i+=1
        time.sleep(3)
        
```

```python
get_tweets(tweet_count=500, text_query="vaccine", since_date="2020-03-11", until_date="2021-03-11")
```

# Read the json data

```python
path_to_json = 'data/'
json_files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]
json_files
```

```python
data = pd.DataFrame()
for file in json_files: 
    df = pd.read_json("data/"+file, lines=True)
    data=data.append(df)
```
